```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
import requests
import numpy as np
from keys import *
```

```python
plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(15,8))
fig.set_facecolor('#5b626e')
ax.set_facecolor('#5b626e')

ax =  sns.barplot(ax= ax, data= df[:10], x='channel_name', y='average_viewers', color='indigo',)
plt.title('Top 10 Avg Viewers of the Week', fontsize=15, color='black')
plt.xlabel('Streamers', fontsize=13, color='black')
plt.ylabel('AVG Viewers', fontsize=13,color='black')
plt.xticks(color='black',fontsize=11)
plt.yticks(color='black')
for i , g in enumerate(df['average_viewers'][:10]):
    plt.text(i-.20, g+1000, f'{np.round(g)}',fontsize=12, color='black')


```
