```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
import requests
import numpy as np
from keys import *
from matplotlib.ticker import FormatStrFormatter
```

```python
df = pd.read_csv('./data_files/Other_Data/Week Directory.csv')
```

```python
df.head()
```

```python
#Replace all str values as int
for column in df.columns[1:-1]:
    df[column] = df[column].str.replace(' ', '').astype(int)
```

```python
import matplotlib


plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(15,8))
fig.set_facecolor('#5b626e')
ax.set_facecolor('#5b626e')

df[:10].plot(ax=ax,kind='line', x='Game', y='Average Viewers', color='darkviolet',marker='.', markersize=20,linewidth=4)
df[:10].plot(ax=ax,kind='bar', x='Game', y='Peak Viewers', color='deepskyblue')

plt.title('Weekly Peak & Average Viewers For Top 10 Directories', fontsize=15, color='black')
plt.xlabel('Directories', fontsize=13, color='black')
plt.ylabel('Peak Viewers', fontsize=13,color='black')
plt.xticks(color='black',fontsize=11,rotation= 45, ha='right')
plt.yticks(color='black')



ax.yaxis.set_major_formatter(matplotlib.ticker.StrMethodFormatter('{x:,.0f}'))

for i , g in enumerate(df['Average Viewers'][:10]):
    plt.text(i-.2, g+15000, '{:,.0f}'.format(g),fontsize=15, color='black')


plt.savefig('./graphs/Weekly Peak & avg Viewers Directory.png')

```
